```python
# Aplicación interactiva sin callback functions

# Importar libreria
from dash import Dash, html, dcc

# Crear la aplicación
app = Dash(__name__)

# Configurar el layout HTML
app.layout = html.Div(
    # Dropdown menu interactivo
    ['Selecciona un color: ',
     dcc.Dropdown(
         id='color-input',
         options=['Rojo', 'Amarillo', 'Azul', 'Verde', 'Negro'],
         value='Azul'
     ),
     ]
)

# Ejecutar la aplicación
if __name__ == '__main__':
    print('Iniciando la aplicación')
    app.run()
else:
    print('Llegó al else')
```
