```python
# Aplicación interactiva con callback functions

# 1. Importar libreria
from dash import Dash, html, dcc
from dash.dependencies import Output, Input
from dash.exceptions import PreventUpdate

# 2. Crear la aplicación
app = Dash(__name__)

# 3. Configurar el layout HTML
app.layout = html.Div(
    # Dropdown menu interactivo
    [
        'Esta aplicación permite al usuario elegir un color a partir de un dropdown y posteriormente mostrarlo como texto',
        html.Br(),
        html.Br(),    
        'Selecciona un color: ',
        dcc.Dropdown(
            id='color-input',
            options=['Rojo', 'Amarillo', 'Azul', 'Verde', 'Negro'],
        ),
        html.Div(id='color-output'),
     ]
)

# 4. Agregar función callback
@app.callback(
        Output('color-output', 'children'), 
        Input('color-input', 'value')
)

def update_output_div(color):
    if not color:
        raise PreventUpdate
    return f'Color selected: {color}'

# 5. Ejecutar la aplicación
if __name__ == '__main__':
    print('Iniciando la aplicación')
    app.run()
else:
    print('Llegó al else')
```

```python
from dash import Dash, html, dcc
from dash.dependencies import Input, Output
from dash.exceptions import PreventUpdate
import plotly.express as px
import seaborn as sns

df = sns.load_dataset('anscombe')

app = Dash(__name__)

app.layout = html.Div([
    html.H3("Anscombe's Quartet"),
    dcc.Dropdown(id='dropdown', options=['I', 'II', 'III', 'IV']),
    dcc.Graph(id='visual',
              figure=px.scatter(
                  df.query('dataset == "II"'), 
                  x='x', 
                  y='y'
                  )
              )
])

if __name__ == '__main__':
    app.run()
```
